# theory

1.  What is a random variable in probability theory?
- n probability theory, a random variable is a variable that takes on different numerical values based on the outcome of a random event or experiment. It connects the outcomes of an experiment to numbers so that we can do calculations.

For example, when you toss a coin, the random variable could be:

1 if it’s heads

0 if it’s tails

2. What are the types of random variables?
- There are two main types of random variables in probability theory:

Discrete Random Variable:

It takes on a countable number of distinct values.

Examples: Number of students in a class, number of heads in 10 coin tosses, number of cars passing by.

Continuous Random Variable:

It can take on any value within a given range (infinite possible values).

Examples: Height of people, weight of a bag, time taken to finish a race.

In simple words, discrete means “countable” and continuous means “measurable.”

3.  What is the difference between discrete and continuous distributions?
- The difference between discrete and continuous distributions is simple:

Discrete distribution:

Deals with variables that have countable values.

Example: Number of students, number of calls received in an hour.

Values are separate and distinct.

Continuous distribution:

Deals with variables that can take any value in a range (infinite possibilities).

Example: Height, weight, temperature.

Values are measured, not counted.

So, discrete = countable, continuous = measurable

4. What are probability distribution functions (PDF)?
- In simple words, a Probability Distribution Function (PDF) is a function that shows how the probability is distributed over different possible values of a continuous random variable.

For a continuous variable, the PDF gives the likelihood of the variable falling within a particular range.

The total area under the PDF curve is always 1 (meaning 100% probability).

Example: Normal distribution has a bell-shaped PDF curve.

So, a PDF helps us understand how likely different values are for a continuous variable.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- A Probability Distribution Function (PDF) and a Cumulative Distribution Function (CDF) are both used with probability, but they show different things:
PDF (Probability Distribution Function) tells you how likely each possible value is.

CDF (Cumulative Distribution Function) tells you the total chance of getting a value up to a certain point.

So, PDF = chance at a value,
CDF = chance up to a value.

6. What is a discrete uniform distribution?
- A discrete uniform distribution is a type of probability distribution where all possible outcomes are equally likely.

For example, when you roll a fair 6-sided die, each number from 1 to 6 has the same chance of coming up (1/6). So the die roll is a discrete uniform distribution.

7. What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution is a simple probability distribution with these key points:

It has only two possible outcomes: success (1) or failure (0).

The probability of success is p, and the probability of failure is 1 − p.

The mean is p and the variance is p(1 − p).

Example: Tossing a coin — heads (1) or tails (0).

8. What is the binomial distribution, and how is it used in probability?
- A binomial distribution shows the probability of getting a certain number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure) and the same probability of success.

For example, tossing a coin 10 times and counting how many times you get heads. It helps in calculating the chance of getting exactly k successes out of n tries.

9. What is the Poisson distribution and where is it applied?
- The Poisson distribution is a probability distribution that shows how likely it is for a certain number of events to happen in a fixed time or space, when the events happen independently and at a constant average rate.

It is often used to model things like the number of phone calls at a call center in an hour, the number of emails you get in a day, or the number of accidents at a traffic intersection in a week.

10. What is a continuous uniform distribution?
- A continuous uniform distribution is a probability distribution where every value within a certain interval is equally likely to occur.

For example, if you pick a random number between 0 and 1, every number in that range has the same chance of being chosen. It’s used when there is no bias toward any value within the range — all outcomes are equally likely.

11. What are the characteristics of a normal distribution?
- A normal distribution is a bell-shaped curve. It is even on both sides, with most values near the middle. The mean, median, and mode are the same. It shows how data is spread out — most values are close to average.

12. What is the standard normal distribution, and why is it important?
- The standard normal distribution is a special normal distribution with a mean of 0 and a standard deviation of 1. It is important because it helps us compare data and find probabilities easily using Z-scores and standard normal tables. Many real-world problems use it for calculations.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) says that if you take many random samples from any population, the average (mean) of those samples will follow a normal distribution — even if the original data is not normal — as long as the sample size is large enough.

This is important because it lets us use normal distribution methods (like confidence intervals and hypothesis testing) for many real-life problems, even when the data isn’t perfectly normal.

14. How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem relates to the normal distribution because it explains why the sampling distribution of the sample mean becomes normal as the sample size grows, no matter what shape the original data has.

In simple words: because of the CLT, when we take many samples and find their means, those means will form a bell-shaped normal distribution. This is why the normal distribution is used so often in statistics.

15. What is the application of Z statistics in hypothesis testing?
- Z statistics is used in hypothesis testing to decide whether to accept or reject a null hypothesis when you know the population variance or have a large sample size.

In simple terms, you calculate a Z score to see how far your sample result is from the population mean. Then you compare it to critical Z values (like ±1.96 for 95% confidence). If your Z score falls in the rejection region, you reject the null hypothesis. It’s commonly used for tests like comparing sample mean to population mean.

16. How do you calculate a Z-score, and what does it represent?
- A Z-score tells you how many standard deviations a data point is from the mean.

You calculate it using this formula:
Z=X-u/sigma

Where:

X = your data point

u (mu) = the population mean

(sigma) = the population standard deviation

What it means:

A Z-score of 0 means the point is exactly at the mean.

A positive Z-score means it’s above the mean.

A negative Z-score means it’s below the mean.

It helps you understand where your value lies in a normal distribution.

17. What are point estimates and interval estimates in statistics?
- In statistics, point estimates and interval estimates are two ways to estimate something about a population:

A point estimate gives you a single best guess. For example, if you take the average (mean) of a sample, that mean is your point estimate of the population mean.

An interval estimate gives you a range of values where you think the true value lies, usually with a level of confidence. For example, a 95% confidence interval for the mean says you’re 95% sure the true mean is within that range.

So, point estimate = single value.
Interval estimate = range of values with a confidence level.

18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals are important in statistics because they show you how sure you are about an estimate and give you a range of possible values for a population parameter.

Instead of just giving one number (like a point estimate), a confidence interval says, “We are reasonably confident the true value lies within this range.” For example, a 95% confidence interval means if you repeated your sampling many times, about 95% of those intervals would contain the true value.

This helps you make better decisions and understand the uncertainty in your results.

19. hat is the relationship between a Z-score and a confidence interval?
- A Z-score and a confidence interval are closely related because Z-scores help determine the range for a confidence interval in a normal distribution.

When you calculate a confidence interval, you use a Z-score to find out how many standard deviations you need to go above and below the mean to capture a certain percentage of the data. For example, for a 95% confidence interval, you use a Z-score of about 1.96.

So, the Z-score tells you how wide your confidence interval should be based on the level of confidence you want.

20. How are Z-scores used to compare different distributions?
- Z-scores are used to compare different distributions by putting values from different datasets on the same standard scale.

A Z-score shows how many standard deviations a value is from its mean. So, no matter what the original units or distributions are, you can see which value is more extreme or unusual compared to its own distribution.

For example, if one test score has a Z-score of 2.0 and another from a different test has a Z-score of 1.5, the first score is more above average for its test than the second one is for its own test.

21. What are the assumptions for applying the Central Limit Theorem?
- The Central Limit Theorem (CLT) works under some basic assumptions:

Random Sampling: The data should come from random samples.

Independence: Each observation should be independent of the others.

Sample Size: The sample size should be large enough (usually n ≥ 30) for the CLT to hold true, especially if the population is not normal.

Finite Variance: The population should have a finite mean and variance.

22. What is the concept of expected value in a probability distribution?
- The expected value in a probability distribution is the average or mean value you expect if you repeat an experiment many times.

It is calculated by multiplying each possible outcome by its probability and then adding them all up.

For example, for a dice roll, the expected value is:
(1×1/6)+(2×1/6)+...+(6×1/6)=3.5

So, the expected value gives you a long-run average outcome of a random variable.

23. How does a probability distribution relate to the expected outcome of a random variable?
- A probability distribution shows all the possible outcomes of a random variable and how likely each outcome is.

The expected outcome (or expected value) is the average value you’d get if you repeated the experiment many times. It’s calculated using the probabilities from the distribution.

So, the probability distribution provides the information needed to find the expected value — they are directly connected!

```python
# Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the discrete uniform distribution
low = 1   # minimum value
high = 7  # maximum value (exclusive, so this gives 1 to 6 like a die)

# Create a discrete uniform distribution object
dist = randint(low, high)

# Possible values
x = np.arange(low, high)

# Compute PMF
pmf = dist.pmf(x)

# Plot PMF
plt.stem(x, pmf)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.xticks(x)
plt.show()
```

```python
#3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.
    x: The outcome (0 or 1)
    p: Probability of success (0 <= p <= 1)
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution.")

# Example usage:
p = 0.7  # Probability of success
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
'''6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution>'''
import numpy as np
import matplotlib.pyplot as plt

# Define a discrete uniform distribution: values 1 to 6 (like a die)
values = np.arange(1, 7)
probability = 1 / len(values)

# PMF: Each value has equal probability
pmf = np.full_like(values, probability, dtype=float)

# CDF: Cumulative sum of the PMF
cdf = np.cumsum(pmf)

# Plot CDF
plt.step(values, cdf, where='post', marker='o')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.grid(True)
plt.show()
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

# Example dataset
data = [10, 12, 14, 15, 18, 21, 22, 23, 25, 27]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Z-scores:", z_scores)

# Plot Z-scores
plt.scatter(range(len(z_scores)), z_scores, color='blue')
plt.axhline(0, color='red', linestyle='--')  # Mean line at zero
plt.xlabel('Data Point Index')
plt.ylabel('Z-score')
plt.title('Z-scores of the Dataset')
plt.grid(True)
plt.show()
```

```python
#10. mplement the Central Limit Theorem (CLT) using Python for a non-normal distribution
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a non-normal population distribution (exponential)
population = np.random.exponential(scale=2.0, size=10000)

# Step 2: Draw multiple samples and compute sample means
sample_means = []
sample_size = 50  # Size of each sample
num_samples = 1000  # Number of samples

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot original population and sampling distribution of means
plt.figure(figsize=(12, 5))

# Plot the original non-normal distribution
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black')
plt.title('Original Population (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the sampling distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, color='salmon', edgecolor='black')
plt.title('Sampling Distribution of Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Example data: sample scores
sample = np.array([85, 90, 88, 92, 87, 85, 89, 90, 91, 86])

# Given population mean
population_mean = 88

# Population standard deviation (assumed known for Z-test)
population_std = 3

# Calculate sample mean
sample_mean = np.mean(sample)

# Calculate sample size
n = len(sample)

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

print(f"P-value: {p_value:.4f}")

# Conclusion
alpha = 0.05  # significance level

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
'''#16.  Create a confidence interval for a dataset using Python and interpret the result'''
import numpy as np
from scipy import stats

# 1. Generate or use your dataset
data = [12, 15, 14, 10, 13, 14, 16, 11, 15, 12]  # Example dataset

# 2. Calculate the sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)
standard_error = sample_std / np.sqrt(n)

# 3. Define your confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# 4. Get the t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# 5. Calculate margin of error
margin_of_error = t_critical * standard_error

# 6. Compute the confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")

# 7. Interpretation
print("\nInterpretation:")
print(f"We are 95% confident that the true mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")
```

```python
#17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy import stats

# 1. Generate data from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

# 2. Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
standard_error = sample_std / np.sqrt(n)

# 3. Confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# 4. Get the z-critical value (use z for normal dist)
z_critical = stats.norm.ppf(1 - alpha/2)

# 5. Calculate margin of error
margin_of_error = z_critical * standard_error

# 6. Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval (95%): [{ci_lower:.2f}, {ci_upper:.2f}]")

# 7. Interpretation
print("\nInterpretation:")
print(f"We are 95% confident that the true mean of the population is between {ci_lower:.2f} and {ci_upper:.2f}.")
```

```python
# 22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution5
import numpy as np

# 1. Generate a normal distribution dataset
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=100)

# 2. Calculate Range
data_range = np.max(data) - np.min(data)

# 3. Calculate IQR (Q3 - Q1)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# 4. Display results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {IQR:.2f}")
```

```python
# 23. Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# 2. Z-score Normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# 3. Visualization
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Normalized data
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=15, color='lightgreen', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# 4. Print mean and std of normalized data
print(f"Mean after normalization: {np.mean(z_scores):.2f}")
print(f"Std deviation after normalization: {np.std(z_scores):.2f}")
```

```python
# 24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution.
import numpy as np
from scipy.stats import skew, kurtosis

# 1. Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# 2. Calculate skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data)  # By default, Fisher’s definition: normal = 0

# 3. Print results
print(f"Skewness: {data_skewness:.4f}")
print(f"Kurtosis: {data_kurtosis:.4f}")
```
